**Упражнение 3**



1. α=1
В общем виде для N=2k:
$S_{2^k}​>1+\frac{k}{2}$​

При k→∞ правая часть стремится к бесконечности. Следовательно, ряд расходится.


2. Случай α=0.5

Здесь мы можем оценить частичную сумму $S_N$​ снизу:
$S_N$​=$1+\frac{1}{\sqrt{2}}​+\frac{1}{\sqrt{3}}​+⋯+\frac{1}{\sqrt{N}}$​

$S_N​>\frac{1}{\sqrt{N}}+\frac{1}{\sqrt{N}}+...​=\frac{N}{\sqrt{N}}​=\sqrt{N}$​


При N→∞ значение $\sqrt{N}$​​→∞. Ряд расходится.


3. Случай α=2

$\frac{1}{N^2}​<\frac{1}{N(N-1)}​​=\frac{1}{N-1}​-\frac{1}{N}​​$

$S_N​<1+1−\frac{1}{N}​​=2−\frac{1}{N}​​<2$

Так как последовательность частичных сумм $S_N$​ монотонно возрастает и ограничена сверху (числом 2), то по теореме Вейерштрасса она имеет конечный предел. Ряд сходится.

```python
def show_seq(an, n0):
    n = np.arange(1, n0+1, 1)
    sums = np.zeros_like(n, np.float32)
    for i in range(0, n0):
        for j in range(i, n0):
            sums[j] += an(i+1)
    plt.plot(n, sums)
    plt.grid()
    plt.axis("equal")
    plt.xlabel("n")
    plt.ylabel("S")
    plt.show()

show_seq(lambda n: 1/np.sqrt(n), 100)
show_seq(lambda n: 1/n, 100)
show_seq(lambda n: 1/n**2, 100)
```

**Упражнение 6**



1)

$\sum_{n=1}^{\infty} \frac{1}{n\sqrt{n+1}\sqrt{n+2}} < \sum_{n=1}^{\infty} \frac{1}{n\sqrt{n}\sqrt{n}} = \sum_{n=1}^{\infty} \frac{1}{n^2}$ - сходится

2)

$\sum_{n=1}^{\infty} \frac{2+\sin{n}}{\sqrt{n}} > \sum_{n=1}^{\infty} \frac{1}{\sqrt{n}} - расходится$

```python
def show_seq(an, bn, n0):
    n = np.arange(1, n0+1, 1)
    sumsa = np.zeros_like(n, np.float32)
    sumsb = np.zeros_like(n, np.float32)
    for i in range(0, n0):
        for j in range(i, n0):
            sumsa[j] += an(i+1)
            sumsb[j] += bn(i+1)
    plt.plot(n, sumsa, "b", label="An")
    plt.plot(n, sumsb, "r", label="Bn")
    plt.grid()
    plt.axis("equal")
    plt.xlabel("n")
    plt.ylabel("S")
    plt.legend()
    plt.show()
    plt.plot(n, an(n), "g", label="an")
    plt.plot(n, bn(n), "y", label="bn")
    plt.grid()
    plt.axis("equal")
    plt.xlabel("n")
    plt.ylabel("An, Bn")
    plt.legend()
    plt.show()

print("А)")
show_seq(lambda n: 1/(n*np.sqrt(n+1)*np.sqrt(n+2)), lambda n: 1/n**2, 20)
print("Bn - сходится -> сходится и An")
print("Б)")
show_seq(lambda n: (2+np.sin(n))/(np.sqrt(n)), lambda n: 1/np.sqrt(n), 20)
print("Bn - расходится -> расходится и An")
```

**Упражнение 7**



```python
def calcSum(func, eps):
    n = Symbol('n')
    a_n = sympify(func)
    
    ratio = a_n.subs(n, n + 1) / a_n
    n0 = 1
    while True:
        q_val = float(ratio.subs(n, n0))
        if q_val < 1:
            break
        n0 += 1
        
    q = float(ratio.subs(n, n0))
    a_func = lambdify(n, a_n, 'numpy')
    
    k = n0
    while True:
        remainder = a_func(k + 1) / (1 - q)
        if remainder < eps:
            break
        k += 1
        
    indices = np.arange(1, k + 1)
    total_sum = np.sum([float(a_func(i)) for i in indices])
    
    return total_sum
```

**Упражнение 9**



```python
def solve_series_task(formula_str, epsilon):
    n = Symbol('n')
    a_func = lambdify(n, sympify(formula_str), modules=['numpy'])
    
    k = 1
    while True:
        next_term = abs(a_func(k + 1))
        if next_term < epsilon:
            break
        k += 1

    indices = np.arange(1, k + 1)
    series_sum = np.sum([float(a_func(float(i))) for i in indices])
    
    return series_sum
```
