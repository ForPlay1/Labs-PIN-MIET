Упражнение 3



```python
x = Symbol('x')
n = 4
f = lambda x: x
a = 1
b = 2

# Левые концы отрезков
def left(a,b,n,f):
    h = (b-a)/n
    E_lefts = np.linspace(a, b-h, n)
    leftsum = sum(np.vectorize(f)(E_lefts))*h
    return leftsum

# Правые концы отрезков
def right(a,b,n,f):
    h = (b-a)/n
    E_rights = np.linspace(a+h, b, n)
    rightsum = sum(np.vectorize(f)(E_rights))*h
    return rightsum

print("Левые концы отрезков:", left(a,b,n,f))
print("Правые концы отрезков:", right(a,b,n,f))
```

**Упражнение 4**



```python
x = Symbol('x')
n = 4
f = lambda x: x
a = 1
b = 2

#  Верхние суммы Дарбу
def up(a,b,n,f):
    h = (b-a)/n
    upsum = 0
    for i in range(0, n-1):
        isum = np.linspace(a + i*h, a + (i+1)*h, 100)
        upsum += h*max(np.vectorize(f)(isum))
    return upsum

# Нижние суммы Дарбу
def low(a,b,n,f):
    h = (b-a)/n
    lowsum = 0
    for i in range(0, n-1):
        isum = np.linspace(a + i*h, a + (i+1)*h, 100)
        lowsum += h*min(np.vectorize(f)(isum))
    return lowsum


print("Верхние суммы Дарбу:", up(a,b,n,f))
print("Нижние суммы Дарбу:", low(a,b,n,f))
```
